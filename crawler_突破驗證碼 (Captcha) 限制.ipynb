```python
#圖片部分的處理
import requests
import shutil #抓圖片的
ss = requests.session()
res = ss.get('https://fbfh.trade.gov.tw/fb/image.do',stream=True,verify=False)
f = open('check.jpg','wb')
shutil.copyfileobj(res.raw,f)
f.close()
```
