```python
import requests

res = requests.get("https://free-proxy-list.net/")
#res.text
```

```python
import re
m = re.findall('\d+\.\d+\.\d+\.\d+:\d+',res.text) #抓出所有符合 *.*.*.*的IP
```

```python
#測試哪個IP是可以使用的  透過 ipify 這個網站 :https://www.ipify.org/ 做測試,若有回應則該IP有效，但因為怕回覆時間太慢，所以設定五秒沒回應就是IP失效
validips = [] #可用的IP放進來
for ip in m:
    try:
        res = requests.get('https://api.ipify.org?format=json',proxies = {'http':ip, 'https':ip}, timeout = 5)
        validips.append({'ip':ip})
    except:
        print('FAIL',ip)
```
